# Оператор условного присваивания

Оператор условного присваивания позволяет присвоить значение переменной в зависимости от выполняемых условий.

Синтаксис оператора условного присваивания:

x = [on_true] if [expression] else [on_false]

```python
numbers = (2, 5, 7, 18, 4, 97, 54, 61, 11)
for item in numbers:
    even_odd = "четное число" if (item % 2 == 0) else "нечетное число"
    print(item, '-', even_odd)
```

# Операторы принадлежности

Операторы принадлежности ***in*** и ***not in*** используются для проверки наличия искомой последовательности в объекте.

```python
# Неполный перечень ближайших к Солнечной системе звезд
near_stars = ['Сириус', 'Процион', 'Альфа Центавра', 'Эпсилон Эридана', 'Тау Кита', '61 Лебедя', 'Звезда Барнарда', 'Звезда Тигардена']

star_1 = 'Тау Кита'
star_2 = 'Альтаир'
star_3 = 'Эпсилон Эридана'
star_4 = 'Бетельгейзе'

print(star_1, 'входит в список:', star_1 in near_stars)
print(star_2, 'входит в список:', star_2 in near_stars)
print(star_3, 'не входит в список:', star_3 not in near_stars)
print(star_4, 'не входит в список:', star_4 not in near_stars)
```

## Копирование списка

Если в списке есть повторяющиеся элементы и от них нужно избавиться, то можно можно применить такой прием:
* сначала преобразовать исходный список в множество с помощью функции set()
* а потом полученное множество обратно преобразовать в список с помощью функции list().

```python
# Добавим с список bright_stars дублирующие элементы
bright_stars.append('Вега')
bright_stars.append('Сириус')
bright_stars.append('Арктур')
print(bright_stars)
```

# Числа

## Числа Фибоначчи

Числа Фибоначчи - последовательность чисел, в которой первые два числа равны 1 и 1,

а каждое последующее число равно сумме двух предыдущих чисел.

Числа Фибоначчи можно встретить в природе, в искусстве, в кодировании.

Отношение соседних чисел Фибоначчи, стремится к золотому сечению.

640px-Fibonacci_Spiral.png

Лямбда-функция для создания чисел Фибоначчи.

```python
fibo = lambda n : n if n <= 1 else fibo(n-1) + fibo(n-2)

k = 24
fibo_nums = []
for i in range(k):
    fibo_nums.append(fibo(i+1))

print(fibo_nums)
```

## Числа Леонардо

Числа Леонардо связаны с числами Фибоначчи через формулу L(n) = 2*F(n+1) - 1, n>=0.

Первые двадцать членов последовательности чисел Леонардо таковы:

1, 1, 3, 5, 9, 15, 25, 41, 67, 109, 177, 287, 465, 753, 1219, 1973, 3193, 5167, 8361, 13529.

Отношение соседних чисел Леонардо, так же, как и соседних чисел Фибоначчи, стремится к золотому сечению.

Лямбда-функция для чисел Леонардо на основе функции для чисел Фибоначчи.

```python
leon = lambda n: 1 if n < 1 else 2*fibo(n+1) - 1

k = 24
leo_nums = []
for i in range(k):
    leo_nums.append(leon(i))

print(leo_nums)
```

## Числа Каталана

Числа Каталана встречаются в комбинаторике.

n-е число Каталана можно определить разными способами, например, как:
* количество разбиений выпуклого (n+2)-угольника на треугольники непересекающимися диагоналями
* количество способов соединения 2n точек на окружности n непересекающимися хордами.

```python
def catalan(n):
    if n >= 2:
        c = ((2 * ((2*n) - 1)) / (n+1)) * catalan(n-1)
        return int(c)
    return 1

k = 20
cat_nums = []
for i in range(k):
    cat_nums.append(catalan(i))

print(cat_nums)
```

## Совершенные числа

Совершенное число — натуральное число, равное сумме всех своих собственных делителей (отличных от самого числа).

Первые семь совершенных чисел: 6, 28, 496, 8128, 33550336, 8589869056, 137438691328.

Функция isperfnum() определяет, является ли заданное число совершенным.

```python
def isperfnum(m):
   total = 0
   for z in range(1,m):
      if(m%z == 0):
         total = total+z
   return total == m

numbers = [3, 6, 12, 21, 28, 64, 100, 496, 777, 1024, 8128]

# С помощью функций filter() и isperfnum() выделим из заданного списка только совершенные числа
perf_nums = filter(isperfnum, numbers)
print(f'Совершенные числа из этого списка: {list(perf_nums)}')
```

## Простые числа

Для вычисления простых чисел существует много разных алгоритмов и функций на их основе.

```python
def isprime_1(num):
    if num> 1:
        for n in range(2,num):
            if (num % n) == 0:
                return False
        return True
    else:
        return False

a, b = 64, 11

print(f'Простое ли число {a}: {isprime_1(a)}')
print(f'Простое ли число {b}: {isprime_1(b)}')
```

```python
def isprime_2(num):
    if num == 1:
       return False
    else:
        for n in range(2,int(num**0.5)+1):
            if num%n==0:
              return False
    return True

a, b = 121, 139

print(f'Простое ли число {a}: {isprime_2(a)}')
print(f'Простое ли число {b}: {isprime_2(b)}')
```

```python
def isprime_3(n):
    if n == 1:
       return False
    if n % 2 == 0:
       return n == 2
    d = 3
    while d * d <= n and n % d != 0:
       d += 2
    return d * d > n

a, b = 127, 187

print(f'Простое ли число {a}: {isprime_3(a)}')
print(f'Простое ли число {b}: {isprime_3(b)}')
```

```python
import math

def isprime_4(n):
    if n <= 1:
        return False
    if n % 2 == 0:
        return n == 2

    max_div = math.floor(math.sqrt(n))
    for i in range(3, 1 + max_div, 2):
        if n % i == 0:
            return False
    return True

a, b = 201, 211

print(f'Простое ли число {a}: {isprime_4(a)}')
print(f'Простое ли число {b}: {isprime_4(b)}')
```

Посмотрим, насколько одинаково все четыре функции выдают результат.

```python
start, stop, step = 1, 100, 1
numbers = list(range(start, stop, step))

print(list(filter(isprime_1, numbers)))
print(list(filter(isprime_2, numbers)))
print(list(filter(isprime_3, numbers)))
print(list(filter(isprime_4, numbers)))
```

## Палиндромные простые числа

Палиндромные простые числа одинаково читаются в обоих направлениях: слева направо и справа налево.

Функция isPalPrime() определяет палиндромные простые числа.

```python
def isPalPrime(n):
    if(str(n) == str(n)[::-1]):
       if n % 2 == 0:
          return n == 2
       d = 3
       while d * d <= n and n % d != 0:
             d += 2
       return d * d > n
    else:
      return False

# Зададим список из чисел для дальнейшего отбора из этоого списка палиндромных простых чисел
start, stop, step = 1, 12000, 1
numbers = list(range(start, stop, step))

# С помощью функций filter() и isPalPrime() отберем из заданного списка палиндромные простые числа
pal_primes = filter(isPalPrime, numbers)
print(list(pal_primes))
```
