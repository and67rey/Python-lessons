# Музыка, речь, звуки природы, звуки из космоса и другие звуки

Звук - это упругие волны, которые распространяются в газообразной, жидкой или твердой среде.

Человеческое ухо воспринимает такие упругие волны с частотой приблизительно от 16 герц до 20 000 герц.

За пределом слышимости человеческого уха находятся:
* ультразвук - упругие волны с более высокой частотой (как правило, более 20 000 герц)
* инфразвук - упругие волны с более низкой частотой (как правило, ниже 16 герц).

Источники звуков бывают самые разные: музыкальные звуки, речь, звуки животных и птиц, звуки природы (гром, ветер и другие), искусственные звуки различных машин и механизмов.

# Форматы звуковых файлов

При цифровой звукозаписи для хранения звуковых данных используются различные форматы аудиофайлов.

Обычно их разделяют на несколько групп:
* форматы без сжатия
* форматы со сжатием и без потерь
* форматы со сжатием и с потерями.

Среди часто встречающихся аудиформатов:
* WAV (Waveform Audio File Format) - как правило, используется для хранения несжатого звука
* MP3 - одним из самых распространённых, популярных и экономичных в потреблении памяти форматов, осуществляет сжатие с потерями
* OGG - осуществляет сжатие с потерями, еще более экономичный, чем MP3 формат.

# Музыкальные звуки

Диапазон музыкальных звуков делится на 9 октав, имеющих собственные названия.

Каждая из октав состоит из 7 ступеней (нот) - от ноты "До" (C) до ноты "Си" (H).

Частоты звучания одинковых нот в соседних октавах отличаются ровно в два раза.

Например, частота ноты "Ля" Первой октавы = 440 Гц, а ноты "Ля" Второй октавы = 880 Гц.

Octaves-notes.png

*Автор схемы октавной системы Sergey Pushkin, по лицензии* [*CY-BA 4.0*](https://creativecommons.org/licenses/by-sa/4.0/)

Подключим библиотеку NumPy с помощью инструкции import.

```python
import numpy as np
```

Создадим и прослушаем звук с частотой 440 Гц, который соответствует ноте "Ля" Первой октавы.

```python
# Зададим частоту сигнала в герцах
freq = 440

# Определим длительность сигнала в секундах
T = 5

# Выберем частоту дискретизации (sampling rate)
sr = 44100

# Создадим массив для временного интервала
t = np.linspace(0, T, int(T*sr), endpoint=False)

# Сформируем чистую звуковую волну на частоте 440 Гц
x = 0.5*np.sin(2*np.pi*freq*t)

# Прослушаем созданный звук - это нота "Ля" Первой октавы
Audio(data = x, rate = sr, autoplay = True)
```

Создадим массивы с частотами для нот всех 9 октав:
* в массиве notes_7 - по 7 нот в каждой октаве (без нот с диезами)
* в массиве notes_12 - по 12 нот в каждой октаве (включая ноты с диезами, например, C#).

Частота ноты "Ля" Первой октавы = 440 Гц.

```python
notes_7 = np.array([])
notes_12 = np.array([])
f0 = 440
n0 = 58
n = 108

for i in range(n):
    k = i + 1 - n0
    m = i + 1
    note = round(440 * 2 ** (k/12), 1)
    if ((m % 12 != 2) and (m % 12 != 4) and (m % 12 != 7) and (m % 12 != 9) and (m % 12 != 11)):
        notes_7 = np.append(notes_7, note)
    notes_12 = np.append(notes_12, note)

notes_7 = notes_7.reshape(9,7)
notes_12 = notes_12.reshape(9,12)
```

Подключим библиотеку Pandas с помощью инструкции import.

```python
import pandas as pd
```

Создадим таблицы типа DataFrame для массивов notes_7 и notes_12 - для удобства отображения на экране.

```python
octaves = ['Субконтроктава', 'Контроктава', 'Большая октава', 'Малая октава', 'Первая октава', 'Вторая октава', 'Третья октава', 'Четвертая октава', 'Пятая октава']
notes7 = ['До', 'Ре', 'Ми', 'Фа', 'Соль', 'Ля', 'Си']
notes12 = ['До', 'До-диез', 'Ре', 'Ре-диез', 'Ми', 'Фа', 'Фа-диез', 'Соль', 'Соль-диез', 'Ля', 'Ля-диез', 'Си']

notes_7_df = pd.DataFrame(notes_7, columns = notes7, index = octaves)
notes_12_df = pd.DataFrame(notes_12, columns = notes12, index = octaves)
notes_7_df
```

```python
notes_12_df
```

Создадим функцию gen_note(), с помощью которой можно прослушать звучание любой выбранной ноты из любой выбранной октавы.

```python
def gen_note(octave, note, system = 0, duration = 0.1, sample_rate = 44100, amp = 1000, offset = 0):
    if system == 0:
       freq = notes_7[octave, note-1]
    else:
       freq = notes_12[octave, note-1]
    N = round(duration*sample_rate)
    ts = np.arange(N)/sample_rate
    ys = amp*np.sin(2*np.pi*freq*ts + offset)
    ys = ys.astype(np.int16)
    return ys
```

Создадим и прослушаем гамму, состоящую из нот Первой и Второй октав.

```python
t = 0.5
s_r = 44100

gamma = [(4,1), (4,2), (4,3), (4,4), (4,5), (4,6), (4,7), (5,1), (5,2), (5,3), (5,4), (5,5), (5,6), (5,7)]
music = np.array([])

for i in range(len(gamma)):
    oct = gamma[i][0]
    note = gamma[i][1]

    x = gen_note(oct, note, duration = t, sample_rate = s_r)
    music = np.append(music, x, axis = 0)

Audio(data=music, rate=s_r, autoplay=True)
```

С помощью следующего кода можно услышать, как звучит любая выбранная нота из любой выбранной октавы.

```python
t = 1
s_r = 44100

oct = int(input("""Введите номер октавы
0 - для Субконтроктавы
1 - для Контроктавы
2 - для Малой октавы
3 - для Большой октавы
4 - для Первой октавы
5 - для Второй октавы
6 - для Третьей октавы
7 - для Четвертой октавы
8 - для Пятой октавы
ваш номер октавы: """))

print('\n')

note = int(input("""Введите номер ноты
1 - "До", 2 - "Ре", 3 - "Ми", 4 - "Фа", 5 - "Соль", 6 - "Ля", 7 - "Си"
ваш номер ноты:  """))

x = gen_note(oct, note, duration = t, sample_rate = s_r)
print('\n')

Audio(data=x, rate=s_r, autoplay=True)
```

## Спектрограммы

Спектрограмма — это визуальный способ представления уровня или “громкости” сигнала во времени на различных частотах, присутствующих в форме волны.

На спектрограмме
* по оси X отображается время звукового сигнала t
* по оси Y отображаются частоты звукового сигнала в момент времени t
* а цветом показывается "громкость" звукового сигнала на различных частотах.  

Для получения спектрограммы звуковой сигнал нужно представить в частотно-временной области с помощью функции stft().

При это применяется оконное преобразование Фурье (short-time Fourier transform, STFT).

Для построения спектрограммы используется функция specshow() из модуля display.

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(20, 8))
specs = librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
cb = plt.colorbar()
```

Если в параметрах функции specshow() указать y_axis = 'log', то можно получить спектрограмму в логарифмической шкале.

Часто это позволяет лучше увидеть детали спектрограммы для более низких частот.

```python
plt.figure(figsize=(20, 8))
specs = librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
cb = plt.colorbar()
```

Сформируем и отобразим на одной картинке две спектрограммы для разных звуковых объектов.

```python
width, height = 20, 8
fig, ax = plt.subplots(1, 2, figsize = (width, height))

X1 = librosa.stft(x1)
Xdb1 = librosa.amplitude_to_db(abs(X1))
specs = librosa.display.specshow(Xdb1, sr=sr1, x_axis='time', y_axis='log', ax = ax[0])

X2 = librosa.stft(x2)
Xdb2 = librosa.amplitude_to_db(abs(X2))
specs = librosa.display.specshow(Xdb2, sr=sr2, x_axis='time', y_axis='log', ax = ax[1])

cb = fig.colorbar(specs, ax=ax)

title_1 = ax[0].set(title='Сигналы от первого искусственного спутника Земли, запущенного СССР 4 октября 1957 г.')
title_2 = ax[1].set(title='Первая запись звуков Марса, сделанная марсоходом Perseverance Mars')
```
