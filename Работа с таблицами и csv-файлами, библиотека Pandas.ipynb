# Словарь с данными о системе TRAPPIST-1

У нас есть несколько словарей с разными данными об экзопланетах системы TRAPPIST-1:
* Словарь с радиусами экзопланет системы TRAPPIST-1 в единицах радиуса Земли
* Словарь с массами экзопланет системы TRAPPIST-1 в единицах массы Земли
* Словарь с величинами больших полуосей экзопланет системы TRAPPIST-1 в астрономических единицах
* Словарь с освещенностями экзопланет системы TRAPPIST-1 в единицах освещенности Земли

Объединим все эти словари в один большой словарь.

TRAPPIST-1-SolarSystemComparison.jpg

```python
# Словарь с радиусами экзопланет системы TRAPPIST-1 в единицах радиуса Земли
R_TRAPPIST = {
    'b' : 1.086,
    'c' : 1.056,
    'd' : 0.772,
    'e' : 0.918,
    'f' : 1.045,
    'g' : 1.127,
    'h' : 0.715,
}

# Словарь с массами экзопланет системы TRAPPIST-1 в единицах массы Земли
M_TRAPPIST = {
    'b' : 0.79,
    'c' : 1.63,
    'd' : 0.33,
    'e' : 0.24,
    'f' : 0.36,
    'g' : 0.566,
    'h' : 0.086,
}

# Словарь с величинами больших полуосей экзопланет системы TRAPPIST-1 в астрономических единицах
D_TRAPPIST = {
    'b' : 0.01111,
    'c' : 0.01522,
    'd' : 0.02145,
    'e' : 0.02818,
    'f' : 0.0371,
    'g' : 0.0451,
    'h' : 0.0596
}

# Словарь с освещенностями экзопланет системы TRAPPIST-1 в единицах освещенности Земли
L_TRAPPIST = {
    'b' : 4.25,
    'c' : 2.27,
    'd' : 1.143,
    'e' : 0.662,
    'f' : 0.382,
    'g' : 0.258,
    'h' : 0.131
}
```

```python
TRAPPIST_1 = {'радиусы': R_TRAPPIST,
              'массы': M_TRAPPIST,
              'большие полуоси': D_TRAPPIST,
              'освещенности': L_TRAPPIST
              }
```

## Выборки данных в двумерных массивах

Если мы хотим извлечь из двумерного массива данные по строкам, можно воспользоваться срезами.

Например, извлечем данные только по экзопланетам от 'd' до 'f'.

```python
TRAPPIST_df['d':'f']
```

Такой же результат можно получить, если сделать срез по номерам строк, в которых находятся данные по экзопланетам от 'd' до 'f'.

При этом надо помнить, что:
* нумерация строк начинается с 0
* и строка с номером правой границы среза не включается в получаемую выборку.

```python
TRAPPIST_df[2:5]
```

Если мы хотим извлечь из двумерного массива данные только из отдельных столбцов, то можно использовать список названий этих столбцов.

Например, для получения данных только из столбцов 'массы' и 'освещенности'.

```python
TRAPPIST_df[['массы', 'освещенности']]
```

Если мы хотим получить данные из столбцов 'массы' и 'освещенности' только для экзопланет от 'd' до 'f'.

```python
TRAPPIST_df['d':'f'][['массы', 'освещенности']]
```

## Индексаторы loc и iloc

С помощью индексатора **loc** можно делать выборки из массива по *явным* индексам - то есть по названиям строк и столбцов.

```python
TRAPPIST_df.loc[:'e', :'массы']
```

С помощью индексатора **iloc** можно делать выборки из массива по *неявным* индексам - то есть по номерам строк и столбцов.

При этом надо помнить, что:
* номера строк и столбцов начинаются с 0
* строка или столбец, соответствующий правой границе среза не будут включены в итоговую выборку.

```python
TRAPPIST_df.iloc[2:6, 1:]
```
